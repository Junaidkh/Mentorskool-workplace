```
import pandas as pd
```

```
# reading the html page using read_html it will save our page in list
d = pd.read_html('http://www.howstat.com/cricket/Statistics/IPL/SeriesMatches.asp?s=2019')

# FInding the length of the list
len(d) 
```

```
d[7] # Gotccha, yay we got our table which we are looking for.
```

```
# Now converting that list of table into dataframe
d1 = pd.DataFrame(d[7])
d1
```

```
# Making first row as Header
new_header = d1.iloc[0]
d1 = d1[1:]
d1.columns = new_header
```

```
d1
```

```
# Changing the column name
d1 = d1.rename(columns={'Ground / Location':'Ground'})
```
